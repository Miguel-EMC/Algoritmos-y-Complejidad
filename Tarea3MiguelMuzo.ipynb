El siguiente codigo Python calcula el numero de terminos necesarios en la serie de Maclaurin de
la funcion arcotangente para garantizar una precision de 10^-10:

```python
import sympy as sp

def error_term(n):
    return (-1)**n / (2**(n+1) * (n+1))

def find_terms(precision):
    n = 0
    while True:
        if abs(error_term(n)) < precision:
            return n
        n += 1

precision = 1e-10
n = find_terms(precision)
print(f"Para una precisiÃ³n de {precision}, se necesitan al menos {n} tÃ©rminos.")

```

El siguiente codigo Python calcula el numero de terminos necesarios en la serie de Maclaurin
de la funcion arcotangente

```python
import sympy as sp

def error_term(n, x):
    return (-1)**n / ((1 + x**2)**(n+1) * (n+1))

def find_terms(precision, x):
    n = 0
    while True:
        if abs(error_term(n, x)) < precision:
            return n
        n += 1

precision = 1e-3
x_values = [1/5, 1/239]

for x in x_values:
    n = find_terms(precision, x)
    print(f"Para x = {x}, se necesitan al menos {n} tÃ©rminos para alcanzar una aproximaciÃ³n de pi dentro de {precision}.")
```

```python
import cmath

def calcular_raices(a, b, c):
    # Calcular el discriminante
    delta = b**2 - 4*a*c

    if delta >= 0:
        # Las raÃ­ces son reales
        if b >= 0:
            x1 = (-b - cmath.sqrt(delta)) / (2*a)
            x2 = (2*c) / (-b - cmath.sqrt(delta))
        else:
            x1 = (2*c) / (-b + cmath.sqrt(delta))
            x2 = (-b + cmath.sqrt(delta)) / (2*a)
    else:
        # Las raÃ­ces son complejas
        x1 = (-b / (2*a)) + (cmath.sqrt(-delta) / (2*a)) * 1j
        x2 = (-b / (2*a)) - (cmath.sqrt(-delta) / (2*a)) * 1j

    return x1, x2

# Ejemplo de uso
a = float(input("Ingrese el coeficiente a: "))
b = float(input("Ingrese el coeficiente b: "))
c = float(input("Ingrese el coeficiente c: "))

x1, x2 = calcular_raices(a, b, c)

print(f"Las raÃ­ces son: x1 = {x1}, x2 = {x2}")
```
